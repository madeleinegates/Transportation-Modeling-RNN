```python
import pandas as pd 
from pandas import read_csv

import keras
import matplotlib.pyplot as plt
import os

import numpy as np
from numpy import argmax

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.losses import BinaryCrossentropy

from matplotlib import pyplot as plt
```

Upload data files for use

```python
train = pd.read_csv('/content/drive/My Drive/train_final.csv')
train.head()
```

```python
dev = pd.read_csv('/content/drive/My Drive/dev_final.csv')
dev.head()
```

```python
test = pd.read_csv('/content/drive/My Drive/test_final.csv')
test.head()
```

Separate output column (transportation mode) and preprocess to correct format

```python
y_train = train['y'].tolist()
y_train = np.array(y_train)
y_train.shape
y_train
```

```python
y_dev = dev['y'].tolist()
y_dev = np.array(y_dev)
y_dev.shape
```

```python
y_test = test['y'].tolist()
y_test = np.array(y_test)
y_test.shape
```

Separate input code, ensure formatted correctly

```python
x_train = train.iloc[:, 10:]
x_train.astype('float32')
```

```python
x_dev = dev.iloc[:, 10:]
x_dev.astype('float32')
```

```python
x_test = test.iloc[:, 10:]
x_test.astype('float32')
```

```python
print(x_train.shape, x_dev.shape, x_test.shape)
```

Build model

```python
model = Sequential()
model.add(Dense(13, activation='relu', kernel_initializer='he_normal', input_shape=(45,)))
model.add(Dense(12, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(8, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(4, activation='softmax'))
```

```python
#opt = tf.keras.optimizers.RMSprop(learning_rate = 0.01, decay = 1e-2)
#opt = tf.keras.optimizers.RMSprop(learning_rate = 0.01, decay = 1e-2, momentum = 0.9)
#opt = tf.keras.optimizers.RMSprop(learning_rate = 0.0001, decay = 1e-2, momentum = 0.9)
#opt = tf.keras.optimizers.SGD(learning_rate = 0.0001, decay = 1e-2, momentum = 0.9)
#opt = tf.keras.optimizers.Adam(learning_rate = 0.0001, decay = 1e-2)
#opt = tf.keras.optimizers.SGD()
#opt = tf.keras.optimizers.RMSprop(lr = 0.001)
opt = tf.keras.optimizers.RMSprop(learning_rate = 0.001, decay = 1e-2, momentum = 0.9)
model.compile(optimizer= opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

```python
loss, accuracy = model.evaluate(x_dev, y_dev, verbose=0)
print(loss, accuracy)
```

Reference for model graph creation code:
https://chrisalbon.com/code/deep_learning/keras/visualize_loss_history/

```python
fittingmodel = model.fit(x_train, y_train, epochs=150, batch_size=32, verbose=1, validation_data = (x_dev, y_dev))
```

Making graph of loss over epochs:

```python
# Make the variables for the training & dev accuracy, variable for number of epochs
training_loss = fittingmodel.history['loss']
dev_loss = fittingmodel.history['val_loss']
epochs = range(1, len(training_loss) + 1)

# Make the plot
plt.plot(epochs, training_loss, 'r--')
plt.plot(epochs, dev_loss, 'b-')
plt.legend(['Training Loss', 'Dev Loss'])
plt.xlabel('Epoch #')
plt.ylabel('Loss')
plt.savefig('nn_loss.png', bbox_inches='tight')
```

Making graph of accuracy over epochs:

```python
# Make the variables for the training & dev accuracy, variable for number of epochs
training_acc = fittingmodel.history['accuracy']
dev_acc = fittingmodel.history['val_accuracy']
epochs = range(1, len(training_acc) + 1)

# Make the plot
plt.plot(epochs, training_acc, 'r--')
plt.plot(epochs, dev_acc, 'b-')
plt.legend(['Training Accuracy', 'Dev Accuracy'])
plt.xlabel('Epoch #')
plt.ylabel('Accuracy')
plt.savefig('nn_accuracy.png', bbox_inches='tight')
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=1)
```

```python
loss, accuracy = model.evaluate(x_dev, y_dev, verbose=0)
print(loss, accuracy)
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose = 0)
print(loss, accuracy)
```
